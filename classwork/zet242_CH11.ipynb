```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

#Define known data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 0, 1, 0, 1])

#Create a cubic spline interpolation of the data
cs = CubicSpline(x, y, bc_type='natural')# Generate new x values for a smooth plot
x_interp = np.linspace(0, 5, 100)
y_interp = cs(x_interp)

#Polynomial Fit (Degree of 3)
p_cubic = np.poly1d(np.polyfit(x, y, 3))
y_poly = p_cubic(x_interp)

#Graphing
plt.scatter(x, y, color='black', zorder=5, label="Original Data Points")
plt.plot(x_interp, y_interp, 'red', label="Cubic Spline")
plt.plot(x_interp, y_poly, 'blue', label="Polynomial Fit (Deg 3)")

plt.title("Cubic Spline vs Polynomial fitting")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

The two methods differ in their accuracy. Namely spline ensures that all of these data points were on the line while the polyfit only follows the rough trend of the data points. This spline effectively shows the oscillating nature of the data that the polyfit does not show. However, this is not always ideal for all data sets due to the almost over-fitting nature spline seems to have.
