```python
import numpy as np
import matplotlib.pyplot as plt

#Parameters
x_start, x_end = 0, 2
y_start = 1
#Different step sizes
h_values = np.arange(0.01, 1.01, 0.05)
# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y
#Storage of run
baseline_y = None
baseline_x = None
```

```python
plt.figure(figsize=(12, 8))

for h in h_values:
    num_steps = int((x_end - x_start) / h)
    x_vals = np.arange(x_start, x_end + h, h)
    y_vals = np.zeros(len(x_vals))
    y_vals[0] = y_start

    # Euler's method
    for i in range(len(x_vals) - 1):
        y_vals[i + 1] = y_vals[i] + h * f(x_vals[i], y_vals[i])

    if baseline_y is None:

        baseline_y = y_vals
        baseline_x = x_vals
        total_error = 0
        print(f"Run h={h:.2f}: Baseline Run")
        print()
    else:
        y_baseline_interp = np.interp(x_vals, baseline_x, baseline_y)
        
        #Errors
        point_errors = np.abs(y_vals - y_baseline_interp)
        total_error = np.sum(point_errors)
        
        print(f"Run h={h:.2f}")
        print(f"errors: {np.round(point_errors, 4)}")
        print(f"Total accumulated error vs baseline: {total_error:.4f}")
        print()

    #Plot
    plt.plot(x_vals, y_vals, label=f"h = {h:.2f} (Err: {total_error:.2f})")
    
#Graphing
plt.xlabel('x')
plt.ylabel('y')
plt.title("Euler's Method: Effect of Step Size (h) on Error")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
