```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
x = np.random.rand(500)
y = np.random.rand(500)
Z = np.linspace(0,500,1)
plt.scatter(x,y)
plt.show()
plt.plot(np.cos(x),y,'b-o')
plt.show()
plt.hist(x, 500)
plt.show()

a = np.linspace(-10, 10, 2000)
b = np.linspace(-10, 10, 2000)
X, Y = np.meshgrid(a, b)
Z = a*b
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='b')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(45, 0)
plt.show()
```
