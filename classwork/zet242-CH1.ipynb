```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python

x = np.random.rand(500)
y = np.random.rand(500)
Z = np.linspace(0,500,500)
z=np.linspace(0,2*np.pi,500)
plt.scatter(x,y)
plt.show()
plt.plot(np.cos(z),Z,'b-o')
plt.show()
plt.hist(x, 500)
plt.show()

a = np.linspace(-50,50,50)
b = np.linspace(-50,50,50)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(a, b)
Q=np.sin(X)*np.cos(Y)
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Q, 50, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(45, 0)
plt.show()
```
