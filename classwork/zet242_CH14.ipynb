```python
import numpy as np
import matplotlib.pyplot as plt

#Use f(x) to make things easier to visualize
def f(x):
    return np.sin(x)
```

```python
#Parameters
a = 0                  #Start of interval
b = 2 * np.pi          #End of interval (one full sine wave)
N = 100                #Number of points
h = (b - a) / (N - 1)  #Step size

#Generate x-values
x = np.linspace(a, b, N)

#True derivative of sin(x)
true_derivative = np.cos(x)

#Approximations

#Central Difference Approximation
central_diff = (f(x + h) - f(x - h)) / (2 * h)
#Forward Difference Approximation
forward_diff = (f(x + h) - f(x)) / h
#Backward Difference Approximation
backward_diff = (f(x) - f(x - h)) / h
```

```python
#Canvas of 4 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Approximating Derivative of f(x) = sin(x)', fontsize=12)

#True Derivative
axs[0, 0].plot(x, true_derivative, 'y--', linewidth=2, label='cos(x)')
axs[0, 0].set_title('True Derivative')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel("f'(x)")
axs[0, 0].legend()
axs[0, 0].grid(True)

#Forward Difference
axs[0, 1].plot(x, forward_diff, 'r--', linewidth=2, label='Forward Difference')
axs[0, 1].set_title('Forward Difference')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel("f'(x)")
axs[0, 1].legend()
axs[0, 1].grid(True)

#Backward Difference
axs[1, 0].plot(x, backward_diff, 'b--', linewidth=2, label='Backward Difference')
axs[1, 0].set_title('Backward Difference')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel("f'(x)")
axs[1, 0].legend()
axs[1, 0].grid(True)

#Central Difference
axs[1, 1].plot(x, central_diff, 'g--', linewidth=2, label='Central Difference')
axs[1, 1].set_title('Central Difference')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel("f'(x)")
axs[1, 1].legend()
axs[1, 1].grid(True)

plt.show()
```
