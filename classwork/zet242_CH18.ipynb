```python
import numpy as np
import matplotlib.pyplot as plt

#Euler Simulation
def euler_method(A, h, t_start, t_end, x_0):
    t_values = np.arange(t_start, t_end + h, h)
    x_values = np.zeros((len(t_values), len(x_0)))
    
    x_values[0] = x_0
    I = np.eye(len(x_0))

    for i in range(1, len(t_values)):
        x_values[i] = np.dot((I + h * A), x_values[i-1])
        
    return t_values, x_values
```

```python
#Parameters
h_values = [0.01, 0.1, 0.5, 1.0, 1.5, 2.0,]  
t_start = 0
t_end = 20
x_0 = [1.0, 0.0]

#Test matrix
A = np.array([[ 0,  1], [-1, -1]])
#Identity matrix
I = np.eye(len(x_0))

plt.figure(figsize = (16,8))

for h in h_values:
    #Stability Check
    M = I + h * A
    eigenvalues = np.linalg.eigvals(M)
    magnitudes = np.abs(eigenvalues)

    #Display
    print(f"--- Step Size (h) = {h} ---")
    print(f"Eigenvalues: {eigenvalues}")
    print(f"Magnitudes: {magnitudes}")

    if np.all(magnitudes < 1.):
        print("Solution is Stable")
    else:
        print("Solution may not be Entirely Stable ;)")
    print(" ")

    #Simulate and Plot
    t, x = euler_method(A, h, t_start, t_end, x_0)
    plt.plot(t, x[:, 0], label=f'h = {h}') 

#Graphing
plt.xlabel('Time (t)')
plt.ylabel('Displacement (y)')
plt.title("Stability Analysis of Euler's Method for a 2nd Order ODE")
plt.legend()
plt.grid(True)
plt.show()
```
