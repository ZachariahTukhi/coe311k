```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def MonteCarlo_pi(a):
    hit = 0
    total = a
    for i in range(a):
        x=random.random()
        y=random.random()
        r=x**2+y**2
        if(r<=1):
            hit+=1
    pi_estimate = (4*hit)/total
    return pi_estimate

array_rel_error = np.zeros(100)
array_true_error = np.zeros(100)
rounds = 0

for p in range(0,100):
    calc_pi = MonteCarlo_pi(1000)
    true_pi = np.pi
    true_error = abs(true_pi-calc_pi)
    rel_error = abs((true_pi-calc_pi)/true_pi)
    
    array_rel_error[p] =rel_error
    array_true_error[p] =true_error
    
```

```python

plt.plot(array_true_error, label='Absolute Error', color='b')
plt.xlabel('Run Index')
plt.ylabel('Absolute Error')
plt.title('Errors of Pi Estimate using the Monte Carlo Method')
plt.grid(True)
plt.legend()

plt.plot(array_rel_error, label='Relative Error', color='g')
plt.xlabel('Run Index')
plt.ylabel('Relative Error')
plt.grid(True)
plt.legend()

plt.show()
```
