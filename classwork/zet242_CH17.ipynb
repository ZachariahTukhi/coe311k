```python
import numpy as np
import matplotlib.pyplot as plt

def run_euler_simulation(h_list, k_val, title):
    plt.figure(figsize=(40, 20))
    x_start = 0
    x_end = 10
    y_0 = 1.0
    v_0 = 0.0
    
    for h in h_list:
        x_values = np.arange(x_start, x_end + h, h)
        y_values = np.zeros(len(x_values))
        v_values = np.zeros(len(x_values))
        
        y_values[0] = y_0
        v_values[0] = v_0
        
        for i in range(1, len(x_values)):
            y_n = y_values[i - 1]
            v_n = v_values[i - 1]
            
            y_values[i] = y_n + h * v_n
            v_values[i] = v_n + h * (-k_val * y_n)
        
        plt.plot(x_values, y_values, label=f'y(x), h={h}')
#         plt.plot(x_values, v_values, linestyle='--', label=f'v(x), h={h}')
    
    plt.xlabel('x')
    plt.ylabel('y and v')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#Play with values
h_vals = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]
run_euler_simulation(h_vals, k_val=1.0, title="Euler Method Stability: Varying h (k=1.0)")

# k = 10 h =0.1
run_euler_simulation([0.01], k_val=10.0, title="Euler Method: k=10, h=0.01")
```

As the value of h increases, the approximation of the oscillation becomes more and more unstable. This is most noteable at h = 0.5, but it can be seen early at h = 0.1. It is clearly showing less and less stable parterns as shown by its deviations from how a spring oscillates, with it adding more and more overshoot to the oscillations, becoming less and less accurate. Thus, more unstable.

An increased k value results in the frequency of oscillations increasing as well as the overshoot happening much sooner and increasing in overshoot sooner.

I do not believe that Euler's method can solve this with a larger h value. Larger and larger h values result in more overshoot with each oscillation. This results in numerical instability, where the solution oscillates erratically or diverges.

```python
import numpy as np
import matplotlib.pyplot as plt

def run_euler_simulation(h_list, k_val, title):
    plt.figure(figsize=(40, 20))
    x_start = 0
    x_end = 10
    y_0 = 1.0
    v_0 = 0.0
    
    for h in h_list:
        x_values = np.arange(x_start, x_end + h, h)
        y_values = np.zeros(len(x_values))
        v_values = np.zeros(len(x_values))
        
        y_values[0] = y_0
        v_values[0] = v_0
        
        for i in range(1, len(x_values)):
            y_n = y_values[i - 1]
            v_n = v_values[i - 1]
            
            y_values[i] = y_n + h * v_n
            v_values[i] = v_n + h * (-k_val * y_n)
        
        plt.plot(x_values, y_values, label=f'y(x), h={h}')
        plt.plot(x_values, v_values, linestyle='--', label=f'v(x), h={h}')
    
    plt.xlabel('x')
    plt.ylabel('y and v')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
    
#Play with values
h_vals = [0.01, 0.02, 0.03, 0.05, 0.1, 0.2, 0.3, 0.5]
run_euler_simulation(h_vals, k_val=1.0, title="Euler Method Stability: Varying h (k=1.0)")

# k = 10 h =0.1
run_euler_simulation([0.01], k_val=10.0, title="Euler Method: k=10, h=0.01")
```
