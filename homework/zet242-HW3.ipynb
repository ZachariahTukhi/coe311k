```python
import numpy as np
import matplotlib.pyplot as plt
import math

def TaylorSin(x, n):
    sin_calc = 0
    for i in range(n):
        sin_count = ((-1)**i) * (x**(2*i+1)) / math.factorial(2*i+1)
        sin_calc += sin_count
    return sin_calc

x = np.arange(0,10.1,.1)
terms = [1,2,3,4]

approx_sin = np.zeros((len(terms), len(x)))
error_sin = np.zeros((len(terms), len(x)))

for i in range(len(terms)):
    for j in range(len(x)):
        approx_value = TaylorSin(x[j], terms[i])
        approx_sin[i][j] = approx_value
        
        sin_actual = np.sin(x[j])
        error_sin[i][j] = abs(approx_value - sin_actual)
```

```python
for i in range(len(terms)):
    plt.plot(x, error_sin[i])
plt.xlabel('x')
plt.ylabel('Error')
plt.title('Taylor Series Approximation of sin(x)')
plt.grid(True)
plt.legend(["1 term", "2 terms", "3 terms", "4 terms"])
plt.show()
```

As more terms are added, the truncation error decreases, making the error smaller. However, at larger values of x, the error grows for all approximations, with this being more apparent at higher polynomials/number of terms. Additionally, after a certain value of x, adding more than two terms in this interval gives only a small improvement, so approximations with two or more terms behave similarly.
