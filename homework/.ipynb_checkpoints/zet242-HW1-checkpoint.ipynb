```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def MonteCarlo_pi(a):
    hit = 0
    total = a
    for i in range(a):
        x=random.random()
        y=random.random()
        r=x**2+y**2
        if(r<=1):
            hit+=1
    pi_estimate = (4*hit)/total
    return pi_estimate

array_true_error = []
rounds = []
counter = 10
for p in range(10,1000,20):
    rounds.append(counter)
    calc_pi = round(MonteCarlo_pi(p),15)
    true_pi = round(np.pi,15)
    true_error = abs(true_pi-calc_pi)
    array_true_error.append(true_error)
    counter = counter+20

    
plt.plot(rounds, array_true_error, label='Absolute Error', color='r')
plt.xlabel('Sample Size')
plt.ylabel('Error')
plt.title('Errors of Pi Estimate using the Monte Carlo Method')
plt.grid(True)
plt.legend()
```
